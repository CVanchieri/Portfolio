## Create a 'Liverpool' football club data set.

```
# import pandas library.
import pandas as pd 

# read in the English Premier League data file.
EPL  = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/LSDS-DataSets/master/EnglishPremierLeagueData/EnglishPremierLeague.csv')
print(EPL.shape)
EPL.head()
```

```
# single out 'Liverpoool' in 'HomeTeam' & 'AwayTeam' colums.
LPFCH =  EPL[EPL['HomeTeam']=='Liverpool']
LPFCA =  EPL[EPL['AwayTeam']=='Liverpool']
```

```
# merge the 2 data sets to create an only 'Liverpool' data set.
LPFC_EPL = pd.concat([LPFCH, LPFCA], sort=False, ignore_index=True)
print(LPFC_EPL.shape)
LPFC_EPL.head()
```

```
# save a copy of the data set.
LPFC_EPL.to_csv('LiverpoolFootballClubData_EPL.csv', index=False)
!head LiverpoolFootballClubData_EPL.csv
```
